```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_country = pd.read_csv('../data/asylum-claims-by-country.csv')
df_country
```

```python
df_country[['Country']] = df_country[['Country']].replace('Total applications', 'Total')
df_country
```

```python
for col in df_country.columns.tolist():
    if col != 'Country':
        df_country[col] = df_country[col].str.replace(',', '').astype(int)
df_country.dtypes
```

```python
df_country.to_csv('../data/asylum-claims-by-country.csv', index=False)
```

```python
df_wait = pd.read_csv('../data/asylum-waiting-periods.csv')
df_wait
```

```python
df_wait.columns = ['Year', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']
df_wait
```

```python
df_wait = df_wait.transpose()
df_wait.columns = df_wait.iloc[0]
df_wait = df_wait.drop('Year')
df_wait
```

```python
for col in df_wait.columns.tolist():
    df_wait[col] = df_wait[col].str.replace(',', '').astype(int)
df_wait.dtypes
```

```python
df_wait.to_csv('../data/asylum-waiting-periods.csv')
```

```python
df_grant = pd.read_csv('../data/asylum-grant-rates.csv')
df_grant
```

```python
int_categories = ['Initial decisions 2024', 'Initial decisions 2025']

for col in int_categories:
    df_grant[col] = df_grant[col].str.replace(',', '').astype(int)
df_grant.dtypes
```

```python
float_categories = ['Grant rate 2024', 'Grant rate 2025']

for col in float_categories:
    df_grant[col] = df_grant[col].str.replace('%', '').astype(int)
df_grant.dtypes
```

```python
df_grant.to_csv('../data/asylum-grant-rates.csv', index=False)
```
