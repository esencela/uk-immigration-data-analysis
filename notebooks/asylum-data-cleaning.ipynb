```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_country = pd.read_csv('../data/asylum-claims-by-country.csv')
df_country
```

```python
df_country[['Country']] = df_country[['Country']].replace('Total applications', 'Total')
df_country
```

```python
for col in df_country.columns.tolist():
    if col != 'Country':
        df_country[col] = df_country[col].str.replace(',', '').astype(int)
df_country.dtypes
```

```python
df_country.to_csv('../data/asylum-claims-by-country.csv')
```
