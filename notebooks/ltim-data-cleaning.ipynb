```python
df_country = pd.read_csv('../data/immigration-by-country.csv')
```

```python
df_country.head()
```

```python
for col in df_country.columns:
    if col == 'Year':
        continue
    if col == 'Nationality':
        continue
    df_country[col] = df_country[col].str.replace(',', '').astype(int)
```

```python
df_country
```

```python
df_country['Nationality'] = df_country['Nationality'].replace('United States of America (USA)', 'USA')
df_country['Nationality'] = df_country['Nationality'].replace('Indian', 'India')
df_country['Nationality'] = df_country['Nationality'].replace('Pakistani', 'Pakistan')
df_country['Nationality'] = df_country['Nationality'].replace('Chinese', 'China')
df_country['Nationality'] = df_country['Nationality'].replace('Nigerian', 'Nigeria')
df_country['Nationality'] = df_country['Nationality'].replace('Ukranian', 'Ukraine')
df_country['Nationality'] = df_country['Nationality'].replace('Zimbabwean', 'Zimbabwe')
df_country['Nationality'] = df_country['Nationality'].replace('Nepalese', 'Nepal')
df_country['Nationality'] = df_country['Nationality'].replace('Ghanaian', 'Ghana')
```

```python
df_country = df_country.rename(columns={'Nationality': 'Country'})
df_country
```

```python
df_country.to_csv('../data/immigration-by-country.csv', index=False)
```

```python
df_reason = pd.read_csv('../data/immigration-by-reason.csv')
```

```python
df_reason
```

```python
df_reason.dropna(inplace=True)
df_reason
```

```python
for col in df_reason.columns:
    if col == 'Year':
        df_reason[col] = df_reason[col].astype(int)
    else:
        df_reason[col] = df_reason[col].str.replace(',', '').astype(int)
df_reason
```

```python
df_reason['Total Dependants'] = df_reason[['Work: Dependant', 'Study: Dependant']].sum(axis=1)
df_reason['Total Humanitarian'] = df_reason[['Humanitarian: BNO', 'Humanitarian: Resettlement', 'Humanitarian: Ukraine']].sum(axis=1)
df_reason
```

```python
df_reason = df_reason.rename(columns={'Total Dependants': 'Dependant', 'Total Humanitarian': 'Humanitarian'})
df_reason
```

```python
df_reason['Year'] = [2019, 2020, 2021, 2022, 2023, 2024]
df_reason
```

```python
df_reason.to_csv('../data/immigration-by-reason.csv', index=False)
```
