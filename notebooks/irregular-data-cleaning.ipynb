```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_boats = pd.read_csv('../data/small-boat-arrivals.csv')
df_boats
```

```python
df_boats.dropna(inplace=True)
df_boats
```

```python
for col in df_boats.columns.tolist():
    if col == 'Year':
        df_boats[col] = df_boats[col].astype(int)
    else:
        df_boats[col] = df_boats[col].str.replace(',', '').astype(int)
df_boats
```

```python
df_boats.to_csv('../data/small-boat-arrivals.csv', index=False)
```

```python
df_nation = pd.read_csv('../data/small-boat-nationality.csv')
df_nation
```

```python
df_nation.drop([9, 19, 21], inplace=True)
df_nation
```

```python
df_nation.dtypes
```

```python
for col in df_nation.columns.tolist():
    if col != 'Nationality' and col != '2018':
        df_nation[col] = df_nation[col].str.replace(',', '').astype(int)
df_nation
```

```python
df_nation.to_csv('../data/small-boat-nationality.csv', index=False)
```
