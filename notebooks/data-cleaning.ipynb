```python
df_country = pd.read_csv('../data/immigration-by-country.csv')
```

```python
df_country.head()
```

```python
for col in df_country.columns:
    if col == 'Year':
        continue
    if col == 'Nationality':
        continue
    df_country[col] = df_country[col].str.replace(',', '').astype(int)
```

```python
df_country
```

```python
df_country['Nationality'] = df_country['Nationality'].replace('United States of America (USA)', 'USA')
df_country['Nationality'] = df_country['Nationality'].replace('Indian', 'India')
df_country['Nationality'] = df_country['Nationality'].replace('Pakistani', 'Pakistan')
df_country['Nationality'] = df_country['Nationality'].replace('Chinese', 'China')
df_country['Nationality'] = df_country['Nationality'].replace('Nigerian', 'Nigeria')
df_country['Nationality'] = df_country['Nationality'].replace('Ukranian', 'Ukraine')
df_country['Nationality'] = df_country['Nationality'].replace('Zimbabwean', 'Zimbabwe')
df_country['Nationality'] = df_country['Nationality'].replace('Nepalese', 'Nepal')
df_country['Nationality'] = df_country['Nationality'].replace('Ghanaian', 'Ghana')
```

```python
df_country.to_csv('../data/immigration-by-country.csv', index=False)
```
